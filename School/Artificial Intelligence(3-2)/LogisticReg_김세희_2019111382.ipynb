```python
#Colab 파일 업로드를 위한 과정
from google.colab import files
import io
import pandas as pd

myfile = files.upload()
```

```python
#csv를 불러오기 위해 pandas import, 외부 데이터 사용
import pandas as pd

#encoding = 'cp949' 한글 깨짐 방지
animal_data = pd.read_csv(io.BytesIO(myfile['data-04-zoo(modify).csv']), nrows = 100, encoding='cp949')
```

```python
#데이터 확인
animal_data
```

```python
#x_data 만들기
x_data = animal_data.__array__()

#y_data 만들기
y_data = np.empty((0,3), int)

#[1, 0, 0] = 포유류, [0, 1, 0] = 조류, [0, 0, 1] = 기타
for rows in range(100):
  if animal_data.loc[rows, 'hair'] == 1:
    y_data = np.append(y_data, np.array([[1, 0, 0]]), axis = 0)

  elif animal_data.loc[rows, 'feathers'] == 1:
    y_data = np.append(y_data, np.array([[0, 1, 0]]), axis = 0)

  else:
    y_data = np.append(y_data, np.array([[0, 0, 1]]), axis = 0)
```

```python
#data check
for i in range(100):
  print('X: ', x_data[i] , 'Y: ', y_data[i])
```

```python
#입력을 위한 placeholder
X = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)
```

```python
#은닉층 3개
W1 = tf.Variable(tf.random_uniform([2, 10], -1. , 1.))
W2 = tf.Variable(tf.random_uniform([10, 20], -1. , 1.))
W3 = tf.Variable(tf.random_uniform([20, 10], -1. , 1.))
W4 = tf.Variable(tf.random_uniform([10, 3], -1. , 1.))

b1 = tf.Variable(tf.zeros([10]))
b2 = tf.Variable(tf.zeros([20]))
b3 = tf.Variable(tf.zeros([10]))
b4 = tf.Variable(tf.zeros([3]))
```

```python
#input -> 은닉층 1층
L1 = tf.add(tf.matmul(X, W1) , b1)
#활성화함수 적용
L1 = tf.nn.relu(L1)

#은닉층 1층 -> 은닉층 2층
L2 = tf.add(tf.matmul(L1, W2) , b2)
L2 = tf.nn.relu(L2)

#은닉층 2층 -> 은닉층 3층
L3 = tf.add(tf.matmul(L2, W3) , b3)
L3 = tf.nn.relu(L3)

#은닉층 3층 -> output
model = tf.add(tf.matmul(L3, W4), b4)
```

```python
#손실 함수 작성
cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(labels = Y, logits = model))

#경사하강법 적용
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 0.01)
train_op = optimizer.minimize(cost)

#세션 초기화
init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)

for step in range(100) :
  sess.run(train_op, feed_dict={X: x_data, Y: y_data})

  if(step + 1) % 10 == 0 :
     print(step + 1, sess.run(cost, feed_dict ={X: x_data, Y: y_data}))

prediction = tf.argmax(model, axis = 1)
target = tf.argmax(Y, axis = 1)
print('예측값: ', sess.run(prediction, feed_dict={X: x_data}))
print('실제값: ', sess.run(target, feed_dict={Y: y_data}))


is_correct = tf.equal(prediction, target)
accuacy = tf.reduce_mean(tf.cast(is_correct, tf.float32))
print('정확도: %.2f' % sess.run(accuacy * 100, feed_dict= {X: x_data, Y: y_data}))
```
