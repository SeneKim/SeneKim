## Test 1 ##

```python
# Test 1

def dice():
  count= 0

  while(count >= 0):
    print("\n주사위를 던지자!")

    a = random.randint(1, 6)
    b = random.randint(1, 6)
    c = random.randint(1, 6)

    count += 1
    if a == b & b == c:
      print(a, b, c)
      print("빙고, 던진 횟수:" + str(count))
      break

    else:
      print(a, b, c)
      print("땡, 다시~")
    
  
```

```python
dice()
```

## Test 2 ##

```python
# Test 2
import time
food = ['피자', '치킨', '스테이크','된장찌개', '바나나', '파스타', '감자튀김']
food.sort()

def recommend():
  print("점심 메뉴를 추천해드립니다")
  print(food)
  time.sleep(1)
  menu = random.randint(0, len(food)-1)
  print("추천음식은 바로바로 --->", food[menu])

  ques = input("추가하고 싶은 음식이 있습니까? y/n")

  while(ques == 'y'):
    new = input("음식 이름(그만 추가하려면 q): ")

    if(new == 'q'):
      break

    food.append(new)

  food.sort()
  print(food) 
  print(food.index(input("찾는 음식은?")) + 1, '번째에 있습니다')


```

```python
recommend()
```

```python
#Test 3
list = []

def lottoFunc():
    lottery = random.randint(1, 45)
    return lottery

print("로또 추첨을 시작합니다!!!")

while True:
  if len(list) == 6:
    break

  lotto_num = lottoFunc() 

  if lotto_num not in list:
    list.append(lotto_num)
  
list.sort() 
print("이번 로또 번호 --->", end = " ")
print(list) 
```
