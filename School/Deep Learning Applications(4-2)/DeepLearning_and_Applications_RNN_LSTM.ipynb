```python
import numpy as np
import pandas as pd
import matplotlib as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense, LSTM
```

```python
f = open('BTC_USD_2019-02-28_2020-02-27-CoinDesk.csv', 'r')
coindesk_data = pd.read_csv(f, header = 0)
seq = coindesk_data[['Closing Price (USD)']].to_numpy() # 종가만 가져옴
```

```python
def seq2dataset(seq, window, horizion):
  X = []
  Y = []
  for i in range(len(seq) - (window +horizion) + 1):
    x = seq[i : (i + window)]
    y = (seq[i + window + horizion - 1])
    X.append(x); Y.append(y)

  return np.array(X), np.array(Y)
```

```python
w = 7
h = 1
X, Y = seq2dataset(seq, w, h)
```

```python
split = int(len(X) * 0.7)
x_train = X[0: split]
y_train = Y[0: split]

x_test = X[split: ]
y_test = Y[split: ]
```

```python
print("훈련 data 크기: ", x_train.shape)
print("테스트 data 크기: ", x_test.shape)
```

```python
model = Sequential()
model.add(SimpleRNN(units = 128, activation = 'relu', input_shape = x_train[0].shape))
model.add(Dense(1))
model.compile(loss = 'mae', optimizer = 'adam', metrics=['mae'])
hist = model.fit(x_train, y_train, epochs = 200, batch_size = 1, validation_data = (x_test, y_test), verbose = 2)
```

```python
ev = model.evaluate(x_test, y_test, verbose = 0)
print("loss function: ", ev[0], "MAE: ", ev[1])

pred = model.predict(x_test)
print("평균 절댓값 백분율 오차(MAPE): ", sum(abs(y_test - pred)/y_test) / len(x_test))
```

```python
model2 = Sequential()
model2.add(LSTM(units = 128, activation = 'relu', input_shape = x_train[0].shape))
model2.add(Dense(1))
model2.compile(loss = 'mae', optimizer = 'adam', metrics=['mae'])
hist2 = model2.fit(x_train, y_train, epochs = 200, batch_size = 1, validation_data = (x_test, y_test), verbose = 2)
```

```python
ev = model2.evaluate(x_test, y_test, verbose = 0)
print("loss function: ", ev[0], "MAE: ", ev[1])

pred = model2.predict(x_test)
print("평균 절댓값 백분율 오차(MAPE): ", sum(abs(y_test - pred)/y_test) / len(x_test))
```

```python
plt.plot(hist.history['mae'])
plt.plot(hist.history['val_mae'])
plt.title('Model mae')
plt.ylabel('Mae')
plt.xlabel('Epoch')
plt.ylim([120, 800])
plt.legend(['Train', 'Validation'], loc = 'best')

plt.grid()
plt.show()

x_range = range(len(y_test))
plt.plot(x_range, y_test)
```
