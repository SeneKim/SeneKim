```python
import sys
import sklearn
import numpy as np
import os

np.random.seed(42)

%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rc('axes', labelsize = 14)
mpl.rc('xtick', labelsize = 12)
mpl.rc('ytick', labelsize = 12)
```

## PCA

```python
angle = np.pi/5
stretch = 5
m = 200

np.random.seed(3)
X = np.random.randn(m, 2) / 10
X = X.dot(np.array([[stretch, 0], [0, 1]]))
X = X.dot([[np.cos(angle), np.sin(angle)], [-np.sin(angle), np.cos(angle)]])
```

```python
X_centered = X - X.mean(axis = 0)
U, s, Vt = np.linalg.svd(X_centered)
c1 = Vt.T[:, 0]
c2 = Vt.T[:, 1]

print(Vt.shape)
print(Vt)
```

```python
plt.figure(figsize=(4, 4))
plt.plot(X[:, 0], X[:, 1], "bo", alpha = 0.5)
plt.axis([-1.4, 1.4, -1.4, 1.4])
plt.arrow(0, 0, c1[0], c1[1], head_width = 0.1, linewidth=5, length_includes_head = True, head_length = 0.1, fc = 'k', ec = 'k')
plt.arrow(0, 0, c2[0], c2[1], head_width = 0.1, linewidth=5, length_includes_head = True, head_length = 0.1, fc = 'k', ec = 'k')
plt.text(c1[0] + 0.1, c1[1] - 0.05, r"$\mathbf{c_1}$", fontsize = 22)
plt.text(c2[0] + 0.1, c2[1], r"$\mathbf{c_2}$", fontsize = 22)
plt.xlabel("$x_1$", fontsize =18)
plt.ylabel("$x_2$", fontsize =18, rotation = 0)
plt.grid(True)

plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=1) # 축 하나만 보겠다. 첫번째 주축으로 transpose
X2D = pca.fit_transform(X) # 1차원으로 펴짐

#차원축소
plt.figure(figsize = (4, 2))
plt.plot([-2, 2], [0, 0], "k-", linewidth = 1)

#Z1 = X2D[:, 0]
plt.plot(X2D[:, 0], np.zeros(m), 'bo', alpha = 0.3)
plt.gca().get_yaxis().set_ticks([])
plt.gca().get_xaxis().set_ticklabels([])
plt.axis([-2, 2, -1, 1])
plt.grid(True)

#차원 복구, 완전하게 복구하지는 못함, 어느정도 손실이 존재한다.
X3D_inv = pca.inverse_transform(X2D)
plt.figure(figsize = (4, 2))
plt.plot(X3D_inv[:, 0], X3D_inv[:, 1], 'bo', alpha = 0.5)
plt.axis([-1.4, 1.4, -1.4, 1.4])

plt.xlabel("$x_1$", fontsize =18)
plt.ylabel("$x_2$", fontsize =18, rotation = 0)
plt.grid(True)

plt.show()
```

```python
from sklearn.datasets import fetch_openml

mnist = fetch_openml('mnist_784', version = 1 , as_frame = False)
mnist.target = mnist.target.astype(np.uint8)
```

```python
from sklearn.model_selection import train_test_split

X = mnist['data']
y = mnist['target']

X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
pca = PCA()
pca.fit(X_train)
cumsum = np.cumsum(pca.explained_variance_ratio_)
d = np.argmax(cumsum >= 0.95) + 1
```

```python
d
```

```python
plt.figure(figsize = (6, 4))
plt.plot(cumsum, linewidth = 3)
plt.axis([0, 400, 0, 1])
plt.xlabel("Dimensions")
plt.ylabel("Explained Variance")
plt.plot([d, d], [0, 0.95], "k:")
plt.plot([0, d], [0.95, 0.95], "k:")
plt.plot(d, 0.95, "ko")
plt.annotate("Elbow", xy = (65, 0.85), xytext = (70, 0.7), arrowprops = dict(arrowstyle="->"), fontsize = 16)
plt.grid(True)
plt.show()
```

## LLE

```python
from sklearn.datasets import make_swiss_roll

X, t = make_swiss_roll(n_samples = 1000, noise = 0.2, random_state = 41)
```

```python
axes = [-11.5, 14, -2, 23, -12, 15]

fig = plt.figure(figsize = (6, 5))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=t, cmap=plt.cm.hot)
ax.view_init(10, -70) #3차원을 조금 더 잘 보려고 약간 회전시키기
ax.set_xlabel("$x_1$", fontsize = 18)
ax.set_ylabel("$x_2$", fontsize = 18)
ax.set_zlabel("$x_3$", fontsize = 18)
ax.set_xlim(axes[0:2])
ax.set_ylim(axes[2:4])
ax.set_xlim(axes[4:6])

plt.show()
```

```python
from sklearn.manifold import LocallyLinearEmbedding

# 3차원을 2차원으로 축소하고 연관성을 같이 계산하기 위한 이웃들을 10개로 잡는다. 2차원으로 임베딩된 특성을 뽑게 된다.
lle = LocallyLinearEmbedding(n_components=2, n_neighbors=10, random_state = 41)
X_reduced = lle.fit_transform(X)
```

```python
plt.title("Unrolled swiss roll using LLE", fontsize = 14)
#Z1= X_reduced[:, 0], Z2 = X_reduced[:, 1]
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=t , cmap = plt.cm.hot)
plt.xlabel("$x_1$", fontsize = 18)
plt.ylabel("$x_2$", fontsize = 18)
plt.axis([-0.065, 0.065, -0.1, 0.12])
plt.grid(True)
plt.show()
```
